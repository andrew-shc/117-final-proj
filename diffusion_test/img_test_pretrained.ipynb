# Generate an Image from a Pretrained Diffusion Model
This notebook demonstrates how to generate an image using a pretrained diffusion model (Stable Diffusion) from Hugging Face Diffusers.

```python
# Import required libraries
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
```

```python
# Load the pretrained Stable Diffusion pipeline
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# Generate an image from a text prompt
prompt = "A fantasy landscape"
negative_prompt = "Alex Li"
guidance_scale = 500

image = pipe(
    prompt,
    # negative_prompt=negative_prompt,
    guidance_scale=guidance_scale
).images[0]

# Display the generated image
image.show()
```
